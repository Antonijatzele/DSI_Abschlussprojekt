```python
import pandas as pd

# Daten einlesen aus Github
# Destatis_Einbürgerungsstatistik_12511-0001 

url = "https://raw.githubusercontent.com/Antonijatzele/DSI_Abschlussprojekt/refs/heads/main/Daten/Destatis_Einb%C3%BCrgerungsstatistik_12511-0001_aufbereitet.csv"
df = pd.read_csv(url)
```

```python
# Liniendiagram
import seaborn as sns
import matplotlib.pyplot as plt

# Seaborn-Stil setzen (optional)
sns.set(style="whitegrid")

# Lineplot zeichnen
sns.lineplot(data=df, x='Jahr', y='Anzahl')

# Achsentitel und Diagrammtitel
plt.title("Entwicklung der Einbürgerungen von Ausländern über die Jahre")
plt.xlabel("Jahr")
plt.ylabel("Anzahl Einbürgerungen")

# Plot anzeigen
plt.show()
```

```python
# Seaborn-Stil setzen
sns.set(style="whitegrid")

# Breitere Plotfläche
plt.figure(figsize=(12, 6))

# Liniendiagramm
sns.lineplot(data=df, x='Jahr', y='Anzahl')

# Diagrammtitel und Achsentitel
plt.title("Entwicklung der Einbürgerungen von Ausländern über die Jahre", fontsize=14)

# X-Achse: 5er Schritte + gedrehte Labels
plt.xticks(ticks=range(df['Jahr'].min(), df['Jahr'].max() + 1, 5), rotation=45)

# Plot anzeigen
plt.tight_layout()
plt.show()
```
