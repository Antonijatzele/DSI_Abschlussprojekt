```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV einlesen (ggf. sep=';' anpassen)
df = pd.read_csv('zusammengefuegt.csv', sep=';')

# Plot erstellen
plt.figure(figsize=(10,6))
plt.plot(df['Jahr'], df['insgesamt_arbeitslose_nachaltersgruppe_unter 25 Jahre'], marker='o')
plt.title('Arbeitslose unter 25 Jahre über die Jahre')
plt.xlabel('Jahr')
plt.ylabel('Anzahl Arbeitslose')
plt.grid(True)
plt.show()
```
