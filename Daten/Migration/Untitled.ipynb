```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Beispiel-Daten
altersreihenfolge = list(range(0, 86))
df_men = pd.Series([1000 - i * 5 for i in altersreihenfolge], index=altersreihenfolge)
df_women = pd.Series([1200 - i * 5 for i in altersreihenfolge], index=altersreihenfolge)

# Farben
male_color = "steelblue"
female_color = "lightcoral"

# Plot
fig, ax = plt.subplots(figsize=(10, 8))

# Männer (negative Werte)
sns.barplot(
    x=-df_men.values,
    y=df_men.index,
    color=male_color,
    label='Männer',
    orient='h',
    ax=ax
)

# Frauen (positive Werte)
sns.barplot(
    x=df_women.values,
    y=df_women.index,
    color=female_color,
    label='Frauen',
    orient='h',
    ax=ax
)

# Achsenticks und Titel
ax.set_xlabel("Bevölkerung")
ax.set_ylabel("Alter")

# X-Achse formatieren
ax.set_xticks([])  # Ticks ausblenden
ax.xaxis.set_tick_params(bottom=False)  # Untere Achsenticks ausblenden

# Legende
ax.legend(loc='upper right')

plt.title("Alterspyramide")
plt.tight_layout()
plt.show()
```
