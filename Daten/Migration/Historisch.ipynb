```python
def titel_codierung(titel):
    dict = {
       'Aufenthaltsrecht nach FreizügG/EU': 'Aufenthaltsrecht nach FreizügG/EU',
       'Vom Erfordernis eines Aufenthaltstitels befreit': 'Andere',
       'Unbefristete Niederlassungserlaubnis': 'Unbefristete Niederlassungserlaubnis',
       'Befristete Aufenthaltserlaubnis, Ausbildung': "Erwerbstätigkeit und Ausbildung",
       'Befristete Aufenthaltserlaubnis, Erwerbstätigkeit': "Erwerbstätigkeit und Ausbildung",
       'Befristete AE, völkerrechtl., human., pol. Gründe': "Völkerrechtl., human., pol. Gründe",
       'Befristete Aufenthaltserlaubnis, familiäre Gründe': "Familiäre Gründe",
       'Befristete AE, besondere Gründe und nationale Visa': "Besondere Gründe und nationale Visa",
       'Antrag auf einen Aufenthaltstitel gestellt': 'Antrag auf einen Aufenthaltstitel gestellt', 
       'Duldung': 'Andere',
       'Aufenthaltsgestattung': 'Andere',
       'Ohne Aufenthaltstitel, Duldung oder Gestattung': 'Ohne Aufenthaltstitel, Duldung oder Gestattung'
    }
    return dict[titel]
    
```

```python
df_stat4 = json2df("12521-0008_Titel_Länder")
df_stat4 = df_stat4[df_stat4["LDRGR1"].isin(["DRITT-EU-28", "EUROPA-EU-28"])]
df_stat4 = df_stat4[df_stat4["RECGL3"] != "REC-AE-08"]
df_stat4["Aufenthaltstitel"] = df_stat4["Ausgewählte Aufenthaltstitel"].apply(titel_codierung)
#df_stat4 = df_stat4[["STAG", "Ländergruppierungen", "Value"]]
df_stat4 = df_stat4.groupby(["STAG", "Aufenthaltstitel"])["Value"].sum().reset_index()
df_stat4.to_csv("historisch_titel.csv")
```
