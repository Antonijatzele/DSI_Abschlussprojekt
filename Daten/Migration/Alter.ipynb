```python
from destasis import json2df
import pandas as pd
```

```python
df_stat = json2df("12521-0003_Alter")
df_stat = df_stat[df_stat["STAG"] == "2024-12-31"]
df_stat = df_stat[df_stat["GES"] != "%TOTAL%"]
df_stat = df_stat[df_stat["ALT102"] != "ALTNN"]


#Bi den Deutschen geht es nur bis 85, daher wird das hier auch gekürzt
def conv_alter(alt):
    nummer = int(alt[3:6])
    if nummer >= 86:
        return 85
    else:
        return nummer

df_stat["ALT"] = df_stat["ALT102"].apply(conv_alter)
df_stat = df_stat.groupby(["Ländergruppierungen", "GES", "ALT"])["Value"].sum().reset_index()
df_stat
```

```python
df_stat.to_csv("alterspyramide.csv")
```
